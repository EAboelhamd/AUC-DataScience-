## What is Deep Learning?

Deep Learning is an umbrella of set of algorithms that mimic the network of neurons in a brain. It is a subset of machine learning. It is called deep learning because it makes use of deep neural networks. Basically, the concept is invented by a startup called (DeepMinds) that has been baught by Google. 

#### Deep learning Network's construction:
The network in deep learning consists of three main layers (as shown below):
1. Input Layer
2. Output Layer
3. Hidden layers (all layers in between, and the word deep means the network join neurons in more than two layers).

```python
from IPython.display import Image
Image("Artificial-Neural-Network-Deep-Learning-model.png" )
```

Deep learning is one of the most powerful and very effective algorithms invented so far. Thanks to the hidden neurons that learn the very basic and smaller features and then connected them together to come up with the corrrect classification. 

Each layer represents a deeper level of knowledge, i.e., the hierarchy of knowledge. A neural network with four layers will learn more complex feature than with that with two layers.

The learning occurs in two phases.
1. Applying a nonlinear transformation of the input and create a statistical model as output.
2. Improving the model with a mathematical method known as derivative.

The neural network repeats these two phases hundreds to thousands of time until it has reached a tolerable level of accuracy. The repeat of this two-phase is called an iteration.

## LSTM Model:

The idea of Long short term memory (LSTM) algorithm is an imporovement for the idea of another deep learning algorithm called (Recurrent Neural Networks(RNN)). 

You can explore the idea of RNN using the below link:
https://www.analyticsvidhya.com/blog/2017/12/fundamentals-of-deep-learning-introduction-to-lstm/

In short, LSTM is capable of solving very large and complex sequence prediction problems. 

The name of long short term memory stems from the idea that the alogrithm holds a memory for keeping or forgeting the states. With LSTMs, the information flows through a mechanism known as cell states. This way, LSTMs can selectively remember or forget things. The information at a particular cell state has three different dependencies, listed below:

1. The previous cell state (i.e. the information that was present in the memory after the previous time step)
2. The previous hidden state (i.e. this is the same as the output of the previous cell)
3. The input at the current time step (i.e. the new information that is being fed in at that moment)

Industries use them to move products around for different processes. LSTMs use this mechanism to move information around.

## LSTM implementation:

Tensorflow and Keras offers built in functions for LSTM implementation .. 

all what is needed is to setup these packages an then import them in your code as below .. 

Kindly follow Reference # 6 (for more details about the below implementation) .. 

We gonna work on a benchmark dataset (imdb dataset) of huge number of words to learn the sequence and similarities in their meaning .. 

Read more about the dataset from the below link 

https://rubikscode.net/2018/03/26/two-ways-to-implement-lstm-network-using-python-with-tensorflow-and-keras/

```python
from keras.preprocessing import sequence 
from keras.models import Sequential 
from keras.layers import Dense, Dropout, Embedding, LSTM 
from keras.datasets import imdb ## the dataset that we gonna work on
```

```python
num_words = 1000 
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=num_words) 
```

```python
X_train
```

```python
y_train
```

Define the input and output layers .. 

```python
X_train = sequence.pad_sequences(X_train, maxlen=200) 
X_test = sequence.pad_sequences(X_test, maxlen=200)
```

Define the structure of LSTM 

```python
# Define network architecture and compile 
model = Sequential() 

model.add(Embedding(num_words, 50, input_length=200)) 

model.add(Dropout(0.2)) 

model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))  

model.add(Dense(250, activation='relu')) # relu activation fun is one of the effective activation functions in deep learning

model.add(Dropout(0.2)) 

model.add(Dense(1, activation='sigmoid')) #sigmoid activation function is used, try another activation fun and observe the results

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])  #compile the model
```

LSTM model fitting .. 

```python
model.fit(X_train, y_train, batch_size=64, epochs=10) ## try to change the number of epochs and batch_size and observe the results
```

Model evaluation .. 

```python
print('\nAccuracy: {}'. format(model.evaluate(X_test, y_test)[1]))
```

The accuracy of the model is 84% .. 
try to change the parameters and observe their effect on the accuracy.

## Convolutional Neural Networks

Another very powerful and effective deep learning algorithm is called (Convolutional neural networks or CNN for short). 
CNN is widly used in Computer Vision and it's cabaple of learning the featureas of the input data and correctly classify it in a very effective way. Mainly it depends on two main steps (Convlusional and Pooling) as shown below .. 

### CNN main steps:
1. Each CNN layer learns filters of increasing complexity.
2. The first layers learn basic feature detection filters: edges, corners, etc
3. The middle layers learn filters that detect parts of objects. For faces, they might learn to respond to eyes, noses, etc
4. The last layers have higher representations: they learn to recognize full objects, in different shapes and positions.
